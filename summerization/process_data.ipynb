```python
import json
import os
import glob
import pandas as pd
```

```python
for t in ["train", "valid", "test"]:
    fns = glob.glob(f'/media/yipeng/Samsung/LLM-paper/datasets/AmaSum/verdict/{t}/*.json')
    df = []
    for fn in fns:
        # read json
        with open(fn, 'r') as f:
            data = json.load(f)
        # get text
        inputs = data['input'].split('Act as an e-commerce expert. Summarize the following product reviews.\n')[1]
        target = data['output']
        df.append([inputs, target])
    df = pd.DataFrame(df, columns=['inputs', 'target'])
    df.to_csv(f'/media/yipeng/Samsung/LLM-paper/datasets/AmaSum/verdict/{t}.csv', index=False)
```
